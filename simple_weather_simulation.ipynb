```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Model 1: Simple Weather Simulation (Sunny/Rainy) with Visualization

# 1. Define the transition probabilities
transitions = {
    'Sunny': {'Sunny': 0.8, 'Rainy': 0.2},
    'Rainy': {'Sunny': 0.5, 'Rainy': 0.5}
}

# 2. Set the initial state
current_state = 'Sunny'

# 3. Set the number of steps (e.g., days) to simulate
num_steps = 20 # Increased steps for a better visualization

# Store the history of states for plotting
state_history = [current_state]

print(f"--- Simple Weather Simulation (Start: {current_state}) ---")
print("Day 0:", current_state)

# 4. Run the simulation loop
for i in range(num_steps):
    possible_next_states = list(transitions[current_state].keys())
    probabilities = list(transitions[current_state].values())

    current_state = random.choices(possible_next_states, weights=probabilities, k=1)[0]
    state_history.append(current_state)
    print(f"Day {i+1}:", current_state)

print("--------------------------------------------------")

# --- Visualization for Model 1 ---
plt.figure(figsize=(10, 4))
plt.plot(state_history, marker='o', linestyle='-', color='skyblue')
plt.title('Markov Chain Weather Simulation Over Days')
plt.xlabel('Day')
plt.ylabel('Weather State')
plt.yticks(['Sunny', 'Rainy']) # Explicitly set y-ticks for states
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
